```python
import pandas as pd

ruta_archivo = '/content/drive/MyDrive/Colab data1/data_af.csv'
df = pd.read_csv(ruta_archivo)
df
```

```python
display(df.corr())

from matplotlib import pyplot as plt
plt.matshow(df.corr())
plt.colorbar()
plt.show()
```

```python
df.plot(kind='scatter', x='rating', y='salary', s=32, alpha=0.3)
df.plot(kind='scatter', x='age', y='salary', s=32, alpha=0.3)
df.plot(kind='scatter', x='experience', y='salary', s=32, alpha=0.3)
df.plot(kind='scatter', x='experience', y='rating', s=32, alpha=0.3)
```

```python
X = df[['rating']]
y = df['salary']

display(X)
display(y)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

display(X_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train, y_train)

print(model.intercept_)
print(model.coef_)

y_test_pred = model.predict(X_test)
result_r2 = r2_score(y_test, y_test_pred)
print(result_r2)
```
