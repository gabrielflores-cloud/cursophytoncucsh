```python
import pandas as pd

ruta_archivo = '/content/drive/MyDrive/Colab data1/applicant_list_7.txt'

nombres_columnas = ['nombre', 'apellido', 'EPH', 'ECH', 'EMA', 'ECS', 'ESP', 'OP1', 'OP2', 'OP3']

df = pd.read_csv(ruta_archivo, sep=' ', names=nombres_columnas)

df
```

```python
print(df.shape)
df.info()
```

```python
display(df['ECH'])
```

```python
display(df['ESP'] + df['EMA'])
```

```python
df["score"] = (df["ESP"] + df["EMA"] + df["ECS"]) / 3
display(df["score"])
```

```python
# Definición de una función para calcular puntaje de admisión
# Formula según el departamento
def calcular_puntaje_admision(linea, departamento):
    puntaje = 0

    if departamento == 'Physics':
        puntaje = (linea["EPH"] + linea["EMA"] + linea["EPH"]) / 3
    elif departamento == "Chemistry":
        puntaje = (linea["ECH"] + linea["EPH"] + linea["ESP"]) / 3
    elif departamento == "Engineering":
        puntaje = (linea["EMA"] + linea["ECS"]  + linea["EPH"] + linea["ESP"]) / 4
    elif departamento == "Biotech":
        puntaje = (linea["EMA"] + linea["ECS"]  + linea["ECH"] + linea["ESP"]) / 4
    elif departamento == "Mathematics":
        puntaje = (linea["EMA"] + linea["ECS"] + linea["EPH"] + linea["ECH"]) / 4
    else:
        return None
    return round(puntaje,2)

df["OP1_score"] = df.apply(lambda x: calcular_puntaje_admision(x, x["OP1"]), axis=1)
df["OP2_score"] = df.apply(lambda x: calcular_puntaje_admision(x, x["OP2"]), axis=1)
df["OP3_score"] = df.apply(lambda x: calcular_puntaje_admision(x, x["OP3"]), axis=1)

display(df)
```

```python
# filtrar el DataFrame para el departamento seleccionado y la opción OP1
aspirantes_departamento = df[(df['OP1'] == "Engineering")]

display(aspirantes_departamento)

```

```python
# Ordenar por OP1_score en orden descendente
aspirantes_ordenados = aspirantes_departamento.sort_values(by=['OP1_score'], ascending=[False])

# Mostrar los resultados
display(aspirantes_ordenados)
```

```python
# Seleccionar los N aspirantes con los puntajes más alto
aspirantes_elegidos = aspirantes_ordenados.head(5)

display(aspirantes_elegidos)
```

```python
# Crear la columna 'nombre_completo' combinando 'nombre' y 'apellido'
aspirantes_elegidos['nombre_completo'] = aspirantes_elegidos['nombre'] + ' ' + aspirantes_elegidos['apellido']

# Crear la tabla con las columnas especificadas
tabla_aspirantes_elegidos = pd.DataFrame({
    'nombre_completo': aspirantes_elegidos['nombre_completo'],
    'departmento': aspirantes_elegidos['OP1'],
    'puntaje': aspirantes_elegidos['OP1_score'],
    'opcion': 'OP1'
})

# Presentar el contenido de la tabla
display(tabla_aspirantes_elegidos)

# Imprimir el número de aspirantes elegidos
# Make sure to use the variable that stores the selected department
departamento_seleccionado = aspirantes_elegidos['OP1'].iloc[0] # Assuming all selected have the same department for OP1
print(f"\nNúmero de aspirantes elegidos para {departamento_seleccionado} (OP1): {len(tabla_aspirantes_elegidos)}")
```
